```python
query = requests.get(link)
# print(query.status_code)
if query.status_code != 200:
    print("Error fetching page")
else:
    content = query.content
print(content)
```

```python
soup = BeautifulSoup(query.text, "html5lib")
```

```python
# >> inspect
```

```python
urls = soup.find_all("a", {"class": "s-item__link"})
# print(urls)
print(len(urls))
```

```python
for url in urls:
    # print(url.txt)
    print(url["href"])
```
